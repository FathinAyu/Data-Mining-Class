```python
data['A2'] = data['A2'].fillna((data['A2'].mean())) 
data['A14'] = data['A14'].fillna((data['A14'].mean())) 
data.isnull().any()
```

```python
data['A4'] = data.fillna(data['A4'].value_counts().index[0]) 
data['A5'] = data.fillna(data['A5'].value_counts().index[0])
data['A6'] = data.fillna(data['A6'].value_counts().index[0])
data['A7'] = data.fillna(data['A7'].value_counts().index[0])
data['A1'] = data.fillna(data['A1'].value_counts().index[0])
data.isnull().any()
```

```python
#encode labels
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['A1'] = LabelEncoder().fit_transform(data['A1']) 
data['A4'] = LabelEncoder().fit_transform(data['A4'])
data['A5'] = LabelEncoder().fit_transform(data['A5'])
data['A6'] = LabelEncoder().fit_transform(data['A6'])
data['A7'] = LabelEncoder().fit_transform(data['A7'])
data['A9'] = LabelEncoder().fit_transform(data['A9'])
data['A10'] = LabelEncoder().fit_transform(data['A10'])
data['A12'] = LabelEncoder().fit_transform(data['A12'])
data['A13'] = LabelEncoder().fit_transform(data['A13'])
data['A16'] = LabelEncoder().fit_transform(data['A16'])
data.head(15)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['A2'] = StandardScaler().fit(data[['A2']]).transform(data[['A2']])
data['A3'] = StandardScaler().fit(data[['A3']]).transform(data[['A3']])
data['A8'] = StandardScaler().fit(data[['A8']]).transform(data[['A8']])
data['A11'] = StandardScaler().fit(data[['A11']]).transform(data[['A11']])
data['A13'] = StandardScaler().fit(data[['A13']]).transform(data[['A13']])
data['A14'] = StandardScaler().fit(data[['A14']]).transform(data[['A14']])
data['A15'] = StandardScaler().fit(data[['A15']]).transform(data[['A15']])
data.head()
```

```python
from sklearn.preprocessing import Normalizer 
```

```python
data['A2'] = Normalizer().fit(data[['A2']]).transform(data[['A2']])
data['A3'] = Normalizer().fit(data[['A3']]).transform(data[['A3']])
data['A8'] = Normalizer().fit(data[['A8']]).transform(data[['A8']])
data['A11'] = Normalizer().fit(data[['A11']]).transform(data[['A11']])
data['A13'] = Normalizer().fit(data[['A13']]).transform(data[['A13']])
data['A14'] = Normalizer().fit(data[['A14']]).transform(data[['A14']])
data['A15'] = Normalizer().fit(data[['A15']]).transform(data[['A15']])
data.head()
```
